```python


np.random.seed(42)
data = pd.DataFrame({
    'time': np.tile(np.arange(1, 11), 30),
    'value': np.random.randn(300).cumsum() + 20,
    'group': np.repeat(['A', 'B', 'C'], 100),
    'subgroup': np.tile(np.repeat(['X', 'Y'], 50), 3),
    'weight': np.random.rand(300) * 10
})


palette = sns.color_palette("viridis", 3)

sns.lineplot(
    data=data, 
    x='time', 
    y='value', 
    hue='group', 
    size='subgroup', 
    style='group', 
    weights='weight',
    palette=palette, 
    hue_order=['C', 'B', 'A'], 
    sizes=(0.5, 2.5), 
    size_order=['Y', 'X'], 
    markers=True, 
    estimator='mean', 
    errorbar=('ci', 95), 
    n_boot=1000, 
    seed=42, 
    sort=True, 
    err_style='band', 
    err_kws={'alpha': 0.2}, 
    legend='auto'
)

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


np.random.seed(10)
data = {
    'left_skewed': np.random.exponential(scale=2, size=1000) * -1 + 10,
    'right_skewed': np.random.exponential(scale=2, size=1000),
    'symmetric': np.random.normal(loc=0, scale=1, size=1000)
}


df = pd.DataFrame(data)

df_melted = df.melt(var_name='Distribution Type', value_name='Value')


plt.figure(figsize=(12, 6))
sns.boxplot(x='Distribution Type', y='Value', data=df_melted)
plt.title('Box Plot of Different Distributions')
plt.show()
```
