```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python

x = np.arange(0, 20, 1)
y = 1 + x**2
X = x.reshape(-1, 1)
print(y)
print(X)
linreg = LinearRegression()
linreg.fit(X, y)
print("Coefficients:", linreg.coef_)
print("Intercept:", linreg.intercept_)
plt.plot(x,y, label="Actual Value");  
plt.plot(x,X@linreg.coef_ + linreg.intercept_, label="Predicted Value");  
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Generate some data
np.random.seed(0)
x = np.sort(np.random.rand(100) * 10)
y = np.sin(x) + np.random.normal(scale=0.5, size=x.shape)

# Linear regression model
linear_model = LinearRegression()
linear_model.fit(x.reshape(-1, 1), y)

# Polynomial regression model (degree 3)
polynomial_model = make_pipeline(PolynomialFeatures(degree=3), LinearRegression())
polynomial_model.fit(x.reshape(-1, 1), y)

# Plot data
plt.scatter(x, y, color='black', label='Data')
plt.plot(x, linear_model.predict(x.reshape(-1, 1)), color='blue', label='Linear fit')
plt.plot(x, polynomial_model.predict(x.reshape(-1, 1)), color='red', label='Polynomial fit')
plt.legend()
plt.show()
```
