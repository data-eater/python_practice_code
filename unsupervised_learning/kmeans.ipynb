```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.drop(columns=['species'],inplace=True)
```

```python
iris.head(10)
```

```python
K=range(1,10)
```

```python
score=[]
for k in K:
    kmeans=KMeans(n_clusters=k,random_state=0)
    kmeans.fit(iris)
    score.append(kmeans.inertia_)
print(score)
```

```python

score=pd.DataFrame(score,index=range(1,len(score)+1))

sns.lineplot(data=score,marker='o')
```
