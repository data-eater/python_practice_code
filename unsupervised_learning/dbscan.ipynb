```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import DBSCAN
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.drop(columns=['species'],inplace=True)
```

```python
iris.head(10)
```

```python
dbscan = DBSCAN(eps=1).fit(iris)
```

```python
dbscan.labels_
```

```python
iris['Cluster'] = dbscan.labels_
```

```python
sns.pairplot(data=iris,hue='Cluster',palette='Set1')
```
