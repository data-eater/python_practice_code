```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Create the figure with the specified mosaic layout
fig, axs = plt.subplot_mosaic([["left", "right_top"], ["left", "right_bottom"]])
fig.figure.set_size_inches(8, 8)
# Plot on each of the subplots
axs["left"].plot([1, 2, 3], [4, 5, 6])
axs["left"].set_title("Left plot")

axs["right_top"].scatter([1, 2, 3], [6, 5, 4], color="r", label="Scatter plot")
axs["right_top"].set_title("Right Top plot")


axs["right_bottom"].bar([1, 2, 3], [7, 8, 9], label="Bar plot")
axs["right_bottom"].set_title("Right Bottom plot")

plt.show()
```

```python
import random
x=np.linspace(1,15,100)
print(x)
y=np.sin(x)
y2=np.sin(x) * random.randint(1,2)
y2
```

```python

plt.figure(figsize=(8,6))
plt.plot(x,y2,'o--', color='purple', lw=0.3, ms=3, label='Component 1')
plt.xlabel(xlabel='Voltages [V]')
plt.ylabel(ylabel="Time [t]")
plt.legend(loc="lower right", fontsize=10)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y2,'o--', color='purple', lw=0.2, ms=3, label='Component 1')
plt.plot(x,y,'gd',ms=3, label='Component 2')
plt.xlabel(xlabel='Voltages [V]')
plt.ylabel(ylabel="Time [t]")
plt.legend(loc="lower right", fontsize=10,ncol=2)
```
